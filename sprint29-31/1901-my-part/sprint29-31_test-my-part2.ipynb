```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

from pyocr_classification import classify_products
```

```python
# Correct labels

labels = ["pocari", "ilohas_normal", "mitsuya_cider", "ilohas_peach", "calpis"]
```

```python
# Initialize the imported class and process the necessary preprocessing

ocr = classify_products()
ocr.preprocessing()
```

## 1_debug.jpg

```python
# Show the image

im = Image.open("detected_images/1_debug.jpg", "r")
plt.imshow(np.array(im))

plt.show()
```

```python
# Validate the class

# Transform the image to numpy
imArray = np.array(im)

# Get an index of the product
index, result = ocr.classify(imArray)

# Output the result
print(index)
if index is not None:
    print(labels[index])

print(result)
```
