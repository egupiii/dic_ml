## 【課題5】用語の調査


以下の２つの単語を知る必要があります。これらを調査してまとめてください。


- 尖度


    - 尖度とは，確率変数の確率密度関数や頻度分布の鋭さを表す数字である．正規分布より尖った分布（データが平均付近に集中し，分布の裾が重い）のときには正の値を，正規分布より扁平な分布（データが平均付近から散らばり，分布の裾が軽い）のときには負の値をとる．正規分布の場合には0になる．


- 歪度


    - 歪度とは，分布の非対称性を示す指標である．「右裾が長い」もしくは「左に偏った」分布のときには正の値を，「左裾が長い」もしくは「右に偏った」分布のときには負の値をとる．左右対称の分布（例えば正規分布）の場合には0になる．

## 【課題6】分布の確認


- seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

```python
# check a distribution

import seaborn as sns

sns.distplot(house_prices.SalePrice)
```

```python
import scipy

"尖度は，{}".format(scipy.stats.kurtosis(house_prices.SalePrice))
```

```python
"歪度は，{}".format(scipy.stats.skew(house_prices.SalePrice))
```

- 目的変数に対して対数変換を行う。

```python
log_sale_price = house_prices.SalePrice.apply(np.log)

log_sale_price
```

- 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

```python
sns.distplot(log_sale_price)
```

```python
"対数変換した後の尖度は，{}".format(scipy.stats.kurtosis(log_sale_price))
```

```python
"対数変換した後の歪度は，{}".format(scipy.stats.skew(log_sale_price))
```

- 対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。


    - 対数正規分布に従う変数の対数をとり，正規分布に従う変数を作ること．対数正規分布に従っていると近似できることを仮定した上で，正規分布ではない分布のデータを正規分布に近づけるために対数変換をすること．

特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。


- 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

<u>補足</u>


重回帰分析などで予測する場合、相関が高すぎる特徴量の組み合わせがあると予測に悪影響が出ることがあります。このような場合の簡単な対処方法として、そういった特徴量の組みを探し出して、１つに絞るということがあります。

#### 【解答】


- 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

```python
corr_mat = house_prices.corr()

sns.clustermap(corr_mat, linewidth=0.1, square=True, cmap="Blues")
```

    - 相関係数の高い関係は，例えば，GrLivAreaとTotRmsAbvGrdや，YearBuiltとGarageYrBltなどである．

- 目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

```python
corr_mat
```

```python
sale_price = []
for _ in corr_mat.SalePrice:
    sale_price.append(_)

sale_price.sort()
sale_price.reverse()
sale_price.pop(0)

top10 = []
for _ in range(10):
    top10.append(sale_price[_])

print(top10)

# corr_mat.nlargest(10, "SalePrice")["SalePrice"]
```

```python
top10_corr_mat = corr_mat.loc[["OverallQual","GrLivArea","GarageCars","GarageArea",
                               "TotalBsmtSF","1stFlrSF","FullBath","TotRmsAbvGrd",
                               "YearBuilt","YearRemodAdd"],["SalePrice"]]

top10_corr_mat
```

```python
sns.heatmap(top10_corr_mat, linewidth=0.1, square=True, cmap="Reds")
```

- 選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。

- 選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

```python
new_house_prices = house_prices.loc[:,["OverallQual","GrLivArea","GarageCars",
                                     "GarageArea","TotalBsmtSF","1stFlrSF","FullBath",
                                     "TotRmsAbvGrd","YearBuilt","YearRemodAdd"]]

new_house_prices
```

```python
new_corr_mat = new_house_prices.corr()

new_corr_mat
```

```python
sns.clustermap(new_corr_mat, linewidth=0.1, square=True, annot=True, cmap="Greens")
```

相関係数が高い組み合わせは，GrLivAreaとTotRmsAbvGrd，TotalBsmtSFと1stFlrSF，GarageCarsとGarageAreaの3つである。
