```python
# get a data set of iris

from sklearn.datasets import load_iris

iris = load_iris()

iris
```

```python
# store explanatory variables of the above data

import pandas as pd

X = pd.DataFrame(iris.data, columns=["sepal_length","sepal_width","petal_length",
                                     "petal_width"])

X
```

```python
# store objective variables of the above data

y = pd.DataFrame(iris.target, columns=["Species"])

y
```

## 【課題6】図の作成


データは可視化して確認することが大切です。図を作成していきましょう。Matplotlibやseabornを使ってください。


- ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）


- boxplotを使いデータの分布を可視化する


- violinplotを使いデータの分布を可視化する


boxplotとviolinplotは似た図を表示します。これらはどう違い、それぞれどのような利点があるのか説明してください。

#### 【解答】


- ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

```python
import matplotlib.pyplot as plt

labels = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

plt.pie(df.iloc[:,0:4].count(), labels=labels, autopct="%.1f%%")

plt.show()
```

- boxplotを使いデータの分布を可視化する

```python
import seaborn as sns

sns.set()
sns.set_style("whitegrid")

sns.boxplot(x="Species", y="sepal_length", data=df)

plt.show()
```

```python
sns.set()
sns.set_style("whitegrid")

sns.boxplot(x="Species", y="sepal_width", data=df)

plt.show()
```

```python
sns.set()
sns.set_style("whitegrid")

sns.boxplot(x="Species", y="petal_length", data=df)

plt.show()
```

```python
sns.set()
sns.set_style("whitegrid")

sns.boxplot(x="Species", y="petal_width", data=df)

plt.show()
```

- violinplotを使いデータの分布を可視化する

```python
sns.set()
sns.set_style("whitegrid")
sns.set_palette("Set3")

sns.violinplot(data=df.iloc[:,0:4], split=True)
sns.despine(offset=10, trim=True)

plt.show()
```

- boxplotもviolinplotも、カテゴリデータを含んだ数値データのばらつきを可視化するのに適している。


- boxplotの、真ん中の線は中央値で、箱の上下はそれぞれ第1・第3四分位点、ヒゲの上下はそれぞれ最大・最小値になる。上下の点は、第1・第3四分位点から判断した場合の「外れ値」を意味している。


- violinplotでは、各数値データを分布として表現する。累積値そのものを可視化しているため、各タイプのデータに複数の山(多峰性)を観測することができる。


- 四分位点と中央値にフォーカスしたい場合はboxplot、分布の形や多峰性を見たい場合はviolinplotが便利かもしれない。

## 【課題7】特徴量間の関係の確認


特徴量間の関係を見ていきます。


- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

```python
# check relationships between each feature value

sns.jointplot(df.sepal_length, df.sepal_width, data=df, kind="scatter")

plt.show()
```

```python
sns.jointplot(df.sepal_length, df.petal_length, df, "reg")

plt.show()
```

```python
sns.jointplot(df.sepal_length, df.petal_width, df, "resid")

plt.show()
```

```python
sns.jointplot(df.sepal_width, df.petal_length, df, "kde")

plt.show()
```

```python
sns.jointplot(df.sepal_width, df.petal_width, df, "hex")

plt.show()
```

```python
sns.jointplot(df.petal_length, df.petal_width, df, color="r")

plt.show()
```

- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
