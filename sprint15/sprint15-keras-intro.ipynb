# Implementation of Logistic Regression

<br />

I am going to create an AND gate by logistic regression.

```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import Adam
```

```python
# Prepare a dataset

x_train = np.array([[0,0],[0,1],[1,0],[1,1]])
y_train = np.array([[0],[0],[0],[1]])
```

# Sequential Model

<br />

tf.keras.models.Sequential


https://www.tensorflow.org/api_docs/python/tf/keras/models/Sequential

<br />

Module: tf.keras.layers


https://www.tensorflow.org/api_docs/python/tf/keras/layers

<br />

tf.keras.layers.Dense


https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense

```python
# Fully connected layer

model = tf.keras.Sequential([tf.keras.layers.Dense(1, activation=tf.nn.sigmoid, input_shape=(2,))])
```

```python
# Check the structure of the created model

model.summary()
```

```python
# Compile

model.compile(loss="binary_crossentropy",
              optimizer=tf.train.AdamOptimizer(learning_rate=0.01),
              metrics=["accuracy"])
```

```python
# Fit

history = model.fit(x_train, y_train,
                    batch_size=1,
                    epochs=1000,
                    verbose=1)
```

```python
# If there is a validation dataset

history = model.fit(x_train, y_train,
                    batch_size=1,
                    epochs=1000,
                    verbose=1,
                    validation_data=(x_train, y_train))
```

```python
# Predict

y_pred_proba = model.predict(x_train)[:,0]

# Change probabilities to 0/1
y_pred = np.where(y_pred_proba>0.5, 1, 0)

print("y_pred_proba", y_pred_proba)
print("y_pred", y_pred)
```

```python
# Evaluate

score = model.evaluate(x_train, y_train, verbose=0)

print("Train loss:", score[0])
print("Train accuracy:", score[1])
```

# Functional API

<br />

tf.keras.models.Model


https://www.tensorflow.org/api_docs/python/tf/keras/models/Model

```python
# Input layer
input_data = tf.keras.layers.Input(shape=(2,))
# Output layer
output = tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)(input_data)

model = tf.keras.Model(inputs=input_data, outputs=output)
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",
              optimizer=tf.train.AdamOptimizer(learning_rate=0.01),
              metrics=["accuracy"])
```

```python
history = model.fit(x_train, y_train,
                    batch_size=1,
                    epochs=1000,
                    verbose=1)
```

### In a Case of Several Layers

<br />

The following shows a neural network of 4 layers.

```python
input_data = tf.keras.layers.Input(shape=(2,))
x = tf.keras.layers.Dense(10, activation=tf.nn.relu)(input_data)
x = tf.keras.layers.Dense(10, activation=tf.nn.relu)(x)
x = tf.keras.layers.Dense(10, activation=tf.nn.relu)(x)
output = tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)(x)

model = tf.keras.Model(inputs=input_data, outputs=output)
```

```python
model.summary()
```

#### tf.keras.layers.concatenate


https://www.tensorflow.org/api_docs/python/tf/keras/layers/concatenate

```python
input_data = tf.keras.layers.Input(shape=(2,))
x = tf.keras.layers.Dense(10, activation=tf.nn.relu)(input_data)
x = tf.keras.layers.Dense(10, activation=tf.nn.relu)(x)
# Ramification
y1 = tf.keras.layers.Dense(10, activation=tf.nn.relu)(x)
y2 = tf.keras.layers.Dense(10, activation=tf.nn.relu)(x)
# Concatenate
z = tf.keras.layers.concatenate([y1, y2])
output = tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)(z)

model = tf.keras.Model(inputs=input_data, outputs=output)
```

```python
model.summary()
```
