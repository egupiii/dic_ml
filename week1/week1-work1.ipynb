```python
import time
import matplotlib.pyplot as plt
%matplotlib inline

thickness = 0.00008


start1 = time.time()
# ここから

folded_thickness1 = thickness * (2 ** 43)
print("厚さ:{:.2f}キロメートル".format(folded_thickness1 / 1000))

# ここまでの時間を計測
elapsed_time1 = time.time() - start1
print("time:{}[s]".format(elapsed_time1))


start2 = time.time()
# ここから

times_of_folding = 2
thickness_list = [thickness]

for i in range(43):
    folded_thickness2 = thickness * times_of_folding
    thickness_list.append(folded_thickness2)
    times_of_folding *= 2

print("厚さ:{:.2f}キロメートル".format(folded_thickness2 / 1000))

# ここまでの時間を計測
elapsed_time2 = time.time() - start2
print("time:{}[s]".format(elapsed_time2))

print("配列の要素数:{}".format(len(thickness_list)))

plt.figure(figsize=(18,6))

plt.subplot(1,2,1)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=15)
plt.plot(thickness_list)

plt.subplot(1,2,2)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thickness_list, color="green")

plt.figure(figsize=(8, 5))
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thickness_list, color="yellow", linewidth=10)

print("""
説明: 計算速度を比較すると、2番目のコードの方が良さそう。
    折り曲げた回数が35回を超えたあたりから、紙の厚さが急激に厚くなっている。
    機械学習の分野でPythonが広く使われている理由のひとつは、機械学習向けのライブラリや
    フレームワークが多く存在するからである。PandasやNumpy、TensorFlowやKerasがPythonの
    上に成り立っており、これらを活用して機械学習のモデルを構築することが出来る。
""")
```
