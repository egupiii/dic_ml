```python
# in 100 days

number_of_given_rice_per_day = []
sum_of_given_rice = []
sum_of_rice = 0
given_rice = 1

for i in range(100):
    number_of_given_rice_per_day.append(given_rice)
    sum_of_rice += given_rice
    sum_of_given_rice.append(sum_of_rice)
    given_rice *= 2
```

```python
# graph of the number of given grains of rice per day

import matplotlib.pyplot as plt
%matplotlib inline

plt.title("the number of given grains of rice per day")
plt.xlabel("the number of days")
plt.ylabel("the number of grains rice")
plt.plot(number_of_given_rice_per_day)
plt.show()
```

```python
# graph of the sum of given grains of rice

plt.title("the sum of given grains of rice")
plt.xlabel("the number of days")
plt.ylabel("the number of grains of rice")
plt.plot(sum_of_given_rice, color="red")
plt.show()
```

```python
# definition of verifying how many grains of rice Shinzaemon Sorori got

import datetime

def compute_sorori_shinzaemon(day=100):
    list_n_grains = []
    list_total_grains = []
    sum_of_rice = 0
    given_rice = 1
    for i in range(day):
        list_n_grains.append(given_rice)
        sum_of_rice += given_rice
        list_total_grains.append(sum_of_rice)
        given_rice *= 2
    return list_n_grains, list_total_grains

day1 = datetime.date(2019,1,5)
day_final = datetime.date(2019,4,30)
ml_days = (day_final-day1).days + 1

list_n_grains, list_total_grains = compute_sorori_shinzaemon(ml_days)
```

```python
# result of the above def in the case of 116 days

print(list_total_grains[-1])
```

```python
# graph of the number of given grains of rice per day

plt.title("the number of given grains of rice per day")
plt.xlabel("the number of days")
plt.ylabel("the number of grains rice")
plt.plot(list_n_grains)
plt.show()
```

```python
# graph of the sum of given grains of rice

plt.title("the sum of given grains of rice")
plt.xlabel("the number of days")
plt.ylabel("the number of grains of rice")
plt.plot(list_total_grains, color="red")
plt.show()
```

```python
# how many days can students be alive by eating given grains of rice

def compute_survivable_days(total_grains, n_students):
    n_grains_of_bowl = 3250 # the number of grains of rice in a bowl
    days = total_grains // ((n_grains_of_bowl*3) * n_students)
    return total_grains, n_students, days

total_grains, n_students, days = compute_survivable_days(list_total_grains[-1], 8)
```
