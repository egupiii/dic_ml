```python
# create an array of a single dimention

import numpy as np

x_list = [0]
val = 0
while val <= 1:
    val += 0.01
    x_list.append(val)

np.array(x_list)
```

```python
# create an array of two dimentions

x_array = np.array(x_list)
trans_x = x_array[:,np.newaxis]

y_list = []
for _ in range(len(x_array)):
    y = x_array[_]/2 + 1
    y_list.append(y)

y_array = np.array(y_list)
trans_y = y_array[:,np.newaxis]

np.append(trans_x, trans_y, axis=1)
```

```python
# compute the slopes

slope_list = []
for _ in range(len(x_array) - 1):
    slope = (y_array[_+1]-y_array[_]) / (x_array[_+1]-x_array[_])
    slope_list.append(slope)

slope_array = np.array(slope_list)

last_slope = ((1.01/2 +1)-(1/2 +1)) / (1.01-1)
np.append(slope_array, last_slope)
```

```python
# in a case of y = x^2

# create an array of two dimentions
y_list2 = []
for _ in range(len(x_array)):
    y = x_array[_] ** 2
    y_list2.append(y)

y_array2 = np.array(y_list2)
trans_y2 = y_array2[:,np.newaxis]

print(np.append(trans_x, trans_y2, axis=1))

# compute the slopes
slope_list2 = []
for _ in range(len(x_array) - 1):
    slope = (y_array2[_+1]-y_array2[_]) / (x_array[_+1]-x_array[_])
    slope_list2.append(slope)

slope_array2 = np.array(slope_list2)

last_slope2 = (1.01**2 - 1**2) / (1.01-1)
final_slope_array2 = np.append(slope_array2, last_slope2)
print(final_slope_array2)

# draw the graph
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("the slope of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(final_slope_array2)
plt.show()
```

## 【課題5】最小値の求め方を説明しよう


勾配降下法で求める。まず、ある点で勾配を求め、その中から最も急で下向きの勾配を選んで、次の点に移る。その点でも勾配を求め、最も急な下向きの勾配を選び、次の点に移る。これを何度か繰り返していくと、最短の経路をたどって、関数の最小値を求めることができる。

## 【課題6】最小値を求めよう


$ y = x^2 + 5x $ の最小値を微分法を使って求める。


$$
y' = 2x + 5
$$


$ y' = 0 $ の時、$ x = -5/2 $


$ x < -5/2 $ の時、傾き( $ y' $ )は負の値を取り、$ x > -5/2 $ の時は、正の値を取る。


よって、$ x = -5/2 $ の時、$ y $ は最小値を取り、その値は


$$
y = (-5/2)^2 + 5 * (-5/2) = -25/4
$$


となる。
