## 【課題】統計基礎用語の説明


- **標準偏差について、分散との違いや関係性、どんな時に使う指標なのかをお答えください。**


    - 標準偏差とは、データや確率変数の散らばり具合を示す値で、データが平均値の周りに集中しているのか、あるいは、平均値から離れているのかを把握する時に使う指標である。標準偏差を求めるには、分散の値の正の平方根を取る。よって、標準偏差の大小は、分散の大小に対応している。標準偏差は、単位の次元がデータと同じため、現実のデータの散らばり具合を求める際には、標準偏差が使われることが多い。一方で、確率分布の散らばり具合を求めるには、分散が用いられることが多い。


- **共分散行列について述べてください。**


    - まず、共分散とは、二次元データの間の関連性を求めるものである。片方のデータが大きくなった時に、もう片方のデータも大きくなれば、共分散の値は大きくなり、もう片方のデータが小さくなれば、共分散の値は小さくなる。また、多次元のデータがあった時に、その変数のそれぞれのペアについて共分散を計算したものが、共分散行列である。


- **1次元の正規分布と多次元の正規分布について述べてください。特にどういう時に使うのかについてお答えください。**


    - 正規分布とは、平均値の付近に集積するようなデータの分布を表した連続的な変数に関する確率分布である。一次元の正規分布は、左右対称で、釣り鐘型の性質を持った分布である。多次元の正規分布では、一次元の場合の平均や分散を一般化したものを考える。

```python
# creating date points

import numpy as np

mean1 = [-3, 0]
mean2 = [0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]]

data1 = np.random.multivariate_normal(mean1, cov, 500)
data2 = np.random.multivariate_normal(mean2, cov, 500)
```

```python
# labeling the above data points

labeled_data1 = np.insert(data1, 2, 0, axis=1)
labeled_data2 = np.insert(data2, 2, 1, axis=1)

concat_data = np.concatenate((labeled_data1, labeled_data2))
```

```python
# verifying whether the labeling succeeds by matplotlib

import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(concat_data[:,0], concat_data[:,1], s=50, c=concat_data[:,2], marker='o')

plt.show()
```
