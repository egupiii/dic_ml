```python
import numpy as np
np.set_printoptions(suppress=True) # 指数表示の禁止を設定する
fuji = np.loadtxt("week2_session_mtfuji_data.csv",delimiter=",", skiprows=1)
```

```python
import pandas as pd
Fuji_df = pd.DataFrame(fuji)
Fuji_df = Fuji_df.rename(columns={0: 'x', 1: 'lat',2: 'lng', 3: 'elevation',4: 'distance'})
Fuji_df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(Fuji_df.elevation)
plt.show
```

## 【問題3】勾配降下法の関数を作成

```python
lr = 0.1
iteration = 100
elevation_list = []
grad_list = []
x = 136
pre_x = x - 1

for _ in range(iteration):
       
    grad = (Fuji_df.elevation[pre_x] - Fuji_df.elevation[x]) / ((pre_x - x) + 1e-10)
    print(grad)
    pre_x = round(x)
    x -= lr * grad
    x = round(x)
         
    elevation_list.append(Fuji_df.elevation[x])
    grad_list.append(grad)

print('Elevation:{a}  Parameter:{b}  Grad:{c}'.format(a=Fuji_df.elevation[x], b=x, c=grad))        
```

```python
plt.figure(figsize=(15, 5)) #グラフ全体領域
plt.subplots_adjust(wspace=0.4, hspace=0.6) #グラフ間余白

fig=25

plt.subplot(1, 2, 1) # グラフ配置
plt.plot(grad_list)
plt.title("Grad",loc='center',fontsize=fig)
plt.xlabel("iteration",fontsize=15)
plt.ylabel("grad",fontsize=15)

plt.subplot(1, 2, 2)
plt.plot(elevation_list)
plt.title("Elevation",loc='center',fontsize=fig)
plt.xlabel("iteration",fontsize=15)
plt.ylabel("elevation",fontsize=15)

plt.show()
```
